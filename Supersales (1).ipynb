```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('SampleSuperstore.csv', encoding= 'unicode_escape')
```

```python
df['Discount'] = df['Discount'].astype('int')
```

```python
df.columns
```

```python
sales_state = df.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Sales')
```

```python
ax = sns.countplot(x = 'Category',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Customer1 = df.groupby(['Category'], as_index=False)['Quantity'].sum().sort_values(by='Category', ascending=False, hue = 'Segment').head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = Customer1, x = 'Category',y= 'Quantity')
```

```python
Customer1 = df.groupby(['Category'], as_index=False)['Quantity'].sum().sort_values(by='Category', ascending=False, hue = 'Segment').head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = Customer1, x = 'Category',y= 'Quantity')
```
